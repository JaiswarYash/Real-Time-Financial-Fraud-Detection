### EDA 

```python
# bar plot
sns.barplot(x=data['Class'].value_counts().index, y=data['Class'].value_counts().values)
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Distribution of Class')
plt.show()
```

```python
# conclusion:
# Observation: Class 0 has 284,315 transactions, Class 1 has only 492.
# Interpretation: The dataset is severely imbalanced at 99.83% vs 0.17%.
# Implication: 
'''1. Model trained on raw date will likely predict Class 0 for all cases, leading to poor fraud detection performance.
which means that we need to apply SMOTE technique to balance the dataset bfore training the model.

2 .use standiard scaler to scale the amount and time column before applying SMOTe Techinque to balance the dataset. 
ensuring  that features such as 'Amount' and 'Time' are on a similar scale to the other features, 
which can improve the performance of the SMOTE algorithm and the subsequent model training.'''
```

```python
# box plot to check for outliers
plt.figure(figsize=(10, 8))
sns.boxplot(data=data, x='Class', y='Amount')

plt.title('Distribution of Transaction Amount by Class')
plt.xlabel('Class (0: Non-Fraud, 1: Fraud)')
plt.ylabel('Amount')
plt.show()
```

```python
# boxplot conclusion:

# Observation: The box plot show that there are some outlier in the amount column for both class 0 and class 1.
# interpretation: the Fraud transactions cluster around a lower amount, while non-fraud transactions have higer value amount which seems to more alligined with customer's behaviour.
# this may indicate that fraudsters tend to make smaller transactions to avoid detection, while non-fraud transactions can vary widely in amount.
# implication:
'''1. Amount alone is not a reliable feature for separating fraud from legitimate transactions because both classes contain high-value outliers.
2. requires careful feature Engineering and use of scaling techniques to mitigate the influence of oulier  and imporove model perforamnce.'''
```

```python
# kde plot to check the distribution of the amount by time
plt.figure(figsize=(10, 8))
sns.kdeplot(data=data, x='Time', hue='Class', fill=True, common_norm=False, alpha=0.5 )
plt.title('Distribution of Transaction Amount by Time and Class')
plt.xlabel('Time')
plt.ylabel('Density')
plt.show()
```

```python
# kdeplot conclusion:
# Observation: The KDE plot shows that the distribution of  transaction for both classes is similar and spread out over the time.
#Interpretation: there is no clear pattern in the distribution of transaction amount over time for either class, suggesting that fraud transactions is extrimely rare and can occur at any time,
# making it difficult to detect based on time alone.
# [1]. For class 0 (non-fraud), the distrbution is Data is spread out which indicate that non-fraud transactions occur consistently throughout the time period.
# [2]. For class 1 (fraud), the distribution is also spread out but with a slightly higher value density at lower time values which may indicate that some fraud transactions occur earlier in the time period, but overall there is no strong temporal pattern.
# Implication:
'''1. Time is not a strong predictor of fraud, and models should not rely heavily on temporal features alone for fraud detection.
2. Fraudulent transactions can occur at any time, so real-time monitoring and detection systems are essential for effective fraud prevention.'''
```

```python
# box-plot
plt.figure(figsize=(10, 10))
sns.boxplot(data=data, x='Class', y='Time')
plt.title('Distribution of Transaction Time by Class')
plt.xlabel('Class (0: Non-Fraud, 1: Fraud)')
plt.ylabel('Time')
plt.show()
```

```python
# box-plot conclusion:
# Observation: The box plot shows that the distribution of transaction time for both classes is similar, with a wide range of values and some outliers.
# Interpretation: There is no clear pattern in the distribution of transaction time for either class, suggesting that fraud transactions can occur at any time, making it difficult to detect based on time alone.
# [1]. For class 0 (non-fraud), the distribution is spread out with some outliers, indicating that non-fraud transactions occur consistently throughout the time period.
# [2]. For class 1 (fraud), the distribution is also spread out with some outliers, which may indicate that fraud transactions can occur at any time, but there is no strong temporal pattern.
# Implication:
'''1. Time is not a strong predictor of fraud, and models should not rely heavily on temporal features alone for fraud detection.
2. Fraudulent transactions can occur at any time, so real-time monitoring and detection systems are essential for effective fraud prevention.'''
```
